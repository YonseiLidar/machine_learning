```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
lr = 1e-1
epoch = int(1e1)
x = 1.2
```

```python
def loss(x):
    return x**2

def compute_gradients(x):
    return 2 * x

def apply_gradients(x, grad, lr):
    return x - lr * grad
```

```python
loss_trace = []
theta_trace = []
theta_val = x

for i in range(epoch):   
    loss_val = loss(theta_val)
    theta_trace.append(theta_val)
    loss_trace.append(loss_val)
    
    grad_val = compute_gradients(theta_val)
    theta_val = apply_gradients(theta_val, grad_val, lr)   
```

```python
x_fig = np.linspace(-1.5, 1.5)
y_fig = loss(x_fig)

plt.plot(x_fig, y_fig)
plt.plot(theta_trace, loss_trace, '--*r')
plt.show()
```
