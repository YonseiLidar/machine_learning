```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from data_loading.generate_data import data_loading
from utils.plot_hyperplane import plot_hyperplane
    
x_train, y_train, x_test, y_test, coeff = data_loading(feature_dim=2)
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
print(coeff)
    
a = LinearRegression()
a.fit(x_train, y_train)
print(a.intercept_, a.coef_)

y_test_pred = a.predict(x_test)
print("mean squared error : ", mean_squared_error(y_test, y_test_pred))
print("R square           : ", r2_score(y_test, y_test_pred))
print("R square           : ", a.score(x_test, y_test))
    
plot_hyperplane(x_test, y_test, y_test_pred, a.intercept_, a.coef_)
```
