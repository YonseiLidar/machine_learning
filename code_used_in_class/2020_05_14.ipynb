```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([848,960,1563,910,1440,1944,904,936,912]).reshape((-1,1)).astype(np.float32)
y = np.array([272,313,342,325,291,351,294,367,279]).reshape((-1,1)).astype(np.float32)
A = np.concatenate([np.ones((x.shape[0],1)), x], axis=1)
theta = np.linalg.inv(A.T@A)@A.T@y
y_pred = A@theta
plt.plot(x,y,'o')
plt.plot(x,y_pred)
plt.show()
```

```python
epoch = 100
lr = 1e-1
theta = np.float32([[0],[np.random.uniform()]])

x = np.array([848,960,1563,910,1440,1944,904,936,912]).reshape((-1,1)).astype(np.float32)
y = np.array([272,313,342,325,291,351,294,367,279]).reshape((-1,1)).astype(np.float32)

# standardization
mu = x.mean()
sigma = x.std()
x = (x-mu)/sigma

A = np.concatenate([np.ones((x.shape[0],1)), x], axis=1)

def loss(theta):
    return ((A@theta-y).T@(A@theta-y))[0,0]

def grad(theta):
    return 2*(A.T@A@theta-A.T@y)

def theta_update(theta):
    gradient = grad(theta)
    return theta - lr*gradient

loss_trace = []
for i in range(epoch):
    theta = theta_update(theta)
    loss_trace.append(loss(theta))

y_pred = A@theta

plt.plot(x,y,'o')
plt.plot(x,y_pred)
plt.show()
```
