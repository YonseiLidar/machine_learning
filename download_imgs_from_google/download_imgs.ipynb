# Create downloads_classes list

```python
import os
root_path = "downloads"
classes = sorted([dirname for dirname in os.listdir(root_path)
                  if os.path.isdir(os.path.join(root_path, dirname))])
print(classes)
```

# Create image_paths dict

```python
from collections import defaultdict

image_paths = defaultdict(list)

for cls in classes:
    image_dir = os.path.join(root_path, cls)
    for filepath in os.listdir(image_dir):
        if filepath.endswith(".jpg"):
            image_paths[cls].append(os.path.join(image_dir, filepath))
```

```python
for cls in classes:
    print(image_paths[cls][0:3])
    image_paths[cls].sort()
    print(image_paths[cls][0:3])
```

```python
import matplotlib.image as mpimg
import numpy as np

n_examples_per_class = 2

for cls in classes:
    for i in range(n_examples_per_class):
        img_path = image_paths[cls][i]
        example_image = mpimg.imread(img_path)
        msg = 'class: {:>10s}, max: {}, min: {}'
        print(msg.format(cls, np.max(example_image), np.min(example_image)))
```

```python
%matplotlib inline
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

n_examples_per_class = 2

for cls in classes:
    fig, axes = plt.subplots(1, 2)
    for i, ax in enumerate(axes):
        img_path = image_paths[cls][i]
        example_image = mpimg.imread(img_path)
        ax.set_title("{}".format(cls))
        ax.imshow(example_image)
        ax.axis("off")
    plt.show()
```
