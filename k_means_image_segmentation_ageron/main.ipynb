```python
import matplotlib.pyplot as plt
from matplotlib.image import imread

img = imread("img/ladybug.png")
print(img.shape)

plt.imshow(img)
plt.show()
```

```python
from sklearn.cluster import KMeans

x = img.reshape(-1, 3) # (426400, 3)
segmentation = KMeans(n_clusters=8, random_state=42)
kmeans = segmentation.fit(x)
# print(kmeans.cluster_centers_)
# print(kmeans.labels_.shape) # (426400,)
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(img.shape)

plt.imshow(segmented_img)
plt.show()
```

```python
from sklearn.cluster import KMeans

x = img.reshape(-1, 3)

for k in range(5, 20, 3):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(x)
    segmented_img = kmeans.cluster_centers_[kmeans.labels_]
    segmented_img = segmented_img.reshape(img.shape)

    plt.imshow(segmented_img)
    plt.title('segmented_img with {} centroids'.format(k))
    plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
from sklearn.cluster import KMeans

# car.png from https://depositphotos.com/201003350/stock-photo-moscow-june-2018-view-windshield.html
img = imread("img/car.png")
print(img.shape)

x = img.reshape(-1, 3)

for k in range(3, 10):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(x)
    segmented_img = kmeans.cluster_centers_[kmeans.labels_]
    segmented_img = segmented_img.reshape(img.shape)

    plt.imshow(segmented_img)
    plt.title('segmented_img with {} centroids'.format(k))
    plt.show()
```
