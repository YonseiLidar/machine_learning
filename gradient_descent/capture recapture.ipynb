```python
import numpy as np
import scipy.special as sc  
import matplotlib.pyplot as plt
```

```python
def prob(n, c, r, t):
    return sc.comb(n-c,r-t) * sc.comb(c, t) / sc.comb(n, r)

def main(c, r, t):
    
    prob_list = []
    for n in range(c+r-t, 10*(c+r-t)): 
        prob_list.append(prob(n, c, r, t))
        
    prob_max = max(prob_list)
    idx = prob_list.index(prob_max)
    mle_n = idx + (c+r-t)
    print(f'MLE n : {mle_n}')
    
    plt.plot(range(c+r-t, 10*(c+r-t)), prob_list)
    plt.plot([mle_n, mle_n], [0, prob_list[idx]], 'o--r')
    plt.show()
    
    return mle_n    
```

```python
if __name__ == '__main__':
    c = 10
    r = 10
    t = 3
    main(c, r, t)
```
