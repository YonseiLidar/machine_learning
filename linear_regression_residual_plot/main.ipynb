```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.linspace(-1,1).reshape((-1,1)) # (50,1)
noise = np.random.normal(0,1,size=x.shape) # (50,1)
noise_level = 0.5
y = (1 + 2*x + 3*x**2 + noise_level * noise).reshape((-1,))

reg = LinearRegression()
reg.fit(x,y)
y_pred = reg.predict(x) 

plt.plot(x,y,'o')
plt.plot(x,y_pred)
plt.show()
```

```python
residue = y - y_pred

plt.plot(x,residue,'o')
plt.show()
```

```python
print(x.shape) # (50,1)
x = np.concatenate((x,x**2),axis=1)
print(x.shape) # (50,2)
```

```python
reg = LinearRegression()
reg.fit(x,y)
y_pred = reg.predict(x) 

plt.plot(x[:,0],y,'o')
plt.plot(x[:,0],y_pred)
plt.show()
```

```python
residue = y - y_pred

plt.plot(x[:,0],residue,'o')
plt.show()
```
