```python
def f(x):
    import numpy as np
    return x**2
```

```python
def grad(x):
    import numpy as np
    return 2*x
```

```python
def parameter_update(x_0, lr, gradient):
    import numpy as np
    return x_0 - lr * gradient 
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
x_0 = 1
lr = 1e-1
x_trace = [x_0]
y_trace = [f(x_0)]
for i in range(10):
    gradient = grad(x_0)
    x_0 = parameter_update(x_0, lr, gradient) 
    x_trace.append(x_0)
    y_trace.append(f(x_0))

import numpy as np
x = np.linspace(-5,5,100)
y = f(x)
plt.plot(x, y)
plt.plot(x_trace, y_trace, 'o-r')
plt.show()
```

```python
import numpy as np

x = np.array([848,960,1563,910,1440,1944,904,936,912]).reshape((-1,1)).astype(np.float32)
y = np.array([272,313,342,325,291,351,294,367,279]).reshape((-1,1)).astype(np.float32)
A = np.concatenate([np.ones((x.shape[0],1)), x], axis=1)
theta = np.linalg.inv(A.T@A)@A.T@y
y_pred = A@theta
plt.plot(x,y,'o')
plt.plot(x,y_pred)
plt.show()
```
