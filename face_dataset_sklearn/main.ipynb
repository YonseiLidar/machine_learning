```python
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt

face = fetch_lfw_people(min_faces_per_person=70, resize=0.9)
print(type(face))
print(dir(face))
print()

# print(face.DESCR)
# print()

print(type(face.data))
print(type(face.images))
print(type(face.target))
print(type(face.target_names))
print()

print(face.target_names)
print()

print(face.data.shape)
print(face.images.shape)
print(face.target.shape)
print()
```

```python
for i in range(10):
    plt.imshow(face.images[i])
    plt.title(face.target_names[face.target[i]])
    plt.show()
```
