```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
print(X.shape)
print(y.shape)
#print(y)

plt.scatter(X[y==0][:,0], X[y==0][:,1], c='r')
plt.scatter(X[y==1][:,0], X[y==1][:,1], c='b')
plt.show()
```
